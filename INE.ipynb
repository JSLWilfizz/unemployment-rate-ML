# Building ANN model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout, LeakyReLU
from sklearn.preprocessing import MinMaxScaler 
from keras.utils.vis_utils import plot_model
from keras.constraints import NonNeg
from tensorflow import keras
from keras.initializers import HeNormal
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from tensorflow_addons.metrics import RSquare
```

```python
df_imputed.columns[0:26]
```

```python
train, test = train_test_split(df_imputed, test_size=0.2, random_state=42) # Spliting Data to 80% Training set And 20% to Testing set

trainsetX =   train.values[:,0:25]
trainsetY =   train.values[:,25:26]
testsetX  =   test.values [:,0:25]
testsetY  =   test.values[:,25:26]
# fit scaler on data
scaler1 = MinMaxScaler(feature_range=(0,1)) # Scale between 0 and 1
scaler2 = MinMaxScaler(feature_range=(0,1)) # Scale between 0 and 1
scaler3 = MinMaxScaler(feature_range=(0,1)) # Scale between 0 and 1
scaler4 = MinMaxScaler(feature_range=(0,1)) # Scale between 0 and 1
scaler1.fit(trainsetX)
scaler2.fit(trainsetY)
scaler3.fit(testsetX)
scaler4.fit(testsetY)
```

```python
trainX = scaler1.transform(trainsetX)  #FT,O,DO TRAINING SET
trainY = scaler2.transform(trainsetY)  #Unemployment rate TRAINING SET
testX  = scaler3.transform(testsetX)  #FT,O,DO TESTING SET
testY  = scaler4.transform(testsetY)  #Unemployment rate TESTING SET
```

Best Model:

```python
model = Sequential()
model.add(Dense(308, input_dim=25, kernel_initializer='HeNormal', activation='relu'))
model.add(Dense(44,  kernel_initializer='HeNormal', activation='relu'))
model.add(Dense(20,  kernel_initializer='HeNormal', activation='relu'))
model.add(Dense(92,  kernel_initializer='HeNormal', activation='relu'))
model.add(Dense(188,  kernel_initializer='HeNormal', activation='relu'))
model.add(Dense(308, kernel_initializer='HeNormal',activation='relu'))
model.add(Dense(116, kernel_initializer='HeNormal',activation='relu'))
model.add(Dense(1,kernel_constraint = NonNeg(), kernel_initializer='normal',activation='linear'))
optimizer = keras.optimizers.Adam(learning_rate=0.001)
model.compile(loss='mse', optimizer=optimizer, metrics=[RSquare()])
history=model.fit(trainX, trainY, epochs=100,batch_size=16, verbose=True)
#plot_model(model, to_file='model_plot.png', show_shapes=True, show_layer_names=True)
val_acc_per_epoch = history.history['loss']
plt.plot(history.history['loss'])
#plt.plot(history.history['val_loss'])
plt.yscale("log")
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
print("Evaluate on test data")
results = model.evaluate(testX, testY, batch_size=256)
print("test loss: ", results)
ynew=model.predict(testX)
```

```python
from keras_tuner import Hyperband

def build_model(hp):
    model=Sequential()
    for i in range(hp.Int('num_of_layers',2,10)):         
      #providing range for number of neurons in hidden layers
       model.add(Dense(units=hp.Int('num_of_neurons'+ str(i),min_value=20,max_value=312,step=12),
                                    activation='relu',kernel_initializer='HeNormal'))
    model.add(Dense(1, kernel_constraint = NonNeg(), activation='linear',kernel_initializer='normal'))    #output layer
    #compiling the model
    model.compile(optimizer=keras.optimizers.Adam(hp.Choice('learning rate',values=[1e-2, 1e-3, 1e-4])),loss='mse',
                  metrics=['mse'])
    return model

tuner=Hyperband(build_model, objective='mse',
                    directory='Hyperband Opt4', project_name= 'Test10_mse_HeNormal_curated_UOX_MDENS_cuttof1mw')
tuner.search(trainX,trainY,epochs=100,validation_data=(testX,testY))
tuner.results_summary()
```
